```python
import numpy as np 
import pandas as pd
import matplotlib as plt 

df=pd.read_csv("/home/scholar/Documents/Documents/AIDE/CSR201 Stats &  EDA/flipkartgargets.csv")
```

```python
print(df.head())
```

```python
# Convert 'Purchase_Date' to datetime format
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])

# Extract the month for monthly trend analysis
df['Month'] = df['Purchase_Date'].dt.to_period('M')

# Group by month and count the purchases
monthly_trends = df.groupby('Month').size()

# Display the results
print(monthly_trends)

# Plot the monthly sales trend
monthly_trends.plot(kind='line', marker='o', title='Sales Trends Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Purchases')
plt.show()
```

```python
# Count the occurrences of each product
product_popularity = df['Product_Name'].value_counts()

# Display the results
print(product_popularity)

# Optional: Plot the results
import matplotlib.pyplot as plt

product_popularity.plot(kind='bar', title='Most Popular Products')
plt.xlabel('Product Name')
plt.ylabel('Number of Purchases')
plt.show()
```

```python
# Group by region and product, then count
region_product_trends = df.groupby(['Customer_Region', 'Product_Name']).size().unstack(fill_value=0)

# Display the results
print(region_product_trends)

# Optional: Plot the data as a heatmap for visualization
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(region_product_trends, annot=True, cmap="YlGnBu", fmt="d")
plt.title("Customer Purchasing Behaviors by Region")
plt.xlabel("Product Name")
plt.ylabel("Customer Region")
plt.show()
```
